## Getting data into Pandas

In this case we will load data from the statsmodels.org library

MTCARS is a traditional dataset used for learning.



```python
mtcars = sm.datasets.get_rdataset("mtcars", "datasets", cache=True).data
df = pd.DataFrame(mtcars)
df['carmodel'] = df.index
print(df)
```

## Preview Data

Use the following data dictionary to understand the attributes available

|Attribute | Description |
|---------|--------------|
|	mpg	|Miles/(US) gallon|
|	cyl|	Number of cylinders|

|drat|	Rear axle ratio|

|qsec|	1/4 mile time|


# Create basic charts

```python
#Pie chart
```

```python
import matplotlib.pyplot as plt

#Pie chart showing the count of cars based on their transmission type with labels
transmission_counts = df['am'].value_counts()

# Adding labels
transmission_labels = {0: 'Automatic(0)', 1: 'Manual(1)'}
labeled_counts = transmission_counts.rename(index=transmission_labels)
#create pie chart
labeled_counts.plot.pie(autopct='%1.1f%%', figsize=(5,5), labels=labeled_counts.index)
plt.title('Distribution of Transmission Types')
plt.show()
```

```python
#Bar chart
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

#Bar chart with different colors for each car model
plt.figure(figsize=(12, 6)) # Adjust figure size for better readability of car model names
sns.barplot(x='carmodel', y='mpg', data=df, hue='carmodel', palette='plasma')
plt.title("MPG by Car Model")
plt.xticks(rotation=90)# Rotate x-axis labels for better readability
plt.tight_layout() # Adjust layout to prevent labels overlapping
plt.show()
```

```python
#Scatter plot
```

```python
#Scatter plot of wt vs mpg
df.plot.scatter(x='wt', y='mpg', title='Weight vs. MPG')
plt.xlabel('Weight (1000 lbs)')
plt.ylabel('Miles per Gallon (mpg)')
plt.show()
```

```python
#Bar chart
```

```python
#Bar chart showing the average mpg for each cyl
avg_mpg_by_cyl = df.groupby('cyl')['mpg'].mean()
avg_mpg_by_cyl.plot(kind='bar', title='Average MPG by Number of Cylinders')
plt.xlabel('Number of Cylinders')
plt.ylabel('Average MPG')
plt.xticks(rotation=0) # Keep x-axis labels horizontal for better readability
plt.show()
```

```python
#Histogram
```

```python
#Histogram of weight
df['wt'].plot(kind='hist', bins=5, title='Distribution of Weight (wt)')
plt.xlabel('Weight (1000 lbs)')
plt.ylabel('Frequency')
plt.show()
```
