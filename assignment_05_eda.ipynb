## 4. Univariate Analysis

Business framing:  

Hotels care about trends like average stay length, customer mix, and pricing. A good EDA starts with understanding single variables and their distribution.

- Select at least 3 individual variables to explore
- Use plots and summary methods (e.g. info(), describe(), etc)  to describe the distribution (hint: we are only looking at the values of one variable, so think of plots you've used in the past that DON'T compare 2 variables.)
- Focus on what matters from a business standpoint (e.g., pricing, cancellations, guest types)

```python
# Your code for univariate analysis (e.g., plots, value counts) 🔧
```

```python
# Create a count plot for the 'is_canceled' variable
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='is_canceled')
plt.title('Distribution of Cancellations (0: Not Canceled, 1: Canceled)')
plt.xlabel('Is Canceled')
plt.ylabel('Number of Bookings')
plt.xticks([0, 1], ['Not Canceled', 'Canceled']) # Label x-axis ticks
plt.show()
```

```python
#summary analysis for is_canceled
# Use info() for the is_canceled variable
print("Info of is_canceled:")
df['is_canceled'].info()

# Use describe() for the is_canceled variable
print("\nDescription of is_canceled:")
display(df['is_canceled'].describe())

# Use value_counts() for the is_canceled variable
print("\nValue Counts of is_canceled:")
display(df['is_canceled'].value_counts())

# Calculate and display the percentage of cancellations
print("\nPercentage of Cancellations:")
display(df['is_canceled'].value_counts(normalize=True) * 100)
```

```python
# Create a count plot for the 'hotel' variable
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='hotel')
plt.title('Distribution of Hotel Types')
plt.xlabel('Hotel Type')
plt.ylabel('Number of Bookings')
plt.show()
```

```python
#summary analysis for hotel type
# Use info() for the hotel variable
print("Info of hotel:")
df['hotel'].info()

# Use describe() for the hotel variable
print("\nDescription of hotel:")
display(df['hotel'].describe())

# Use value_counts() for the hotel variable
print("\nValue Counts of hotel:")
display(df['hotel'].value_counts())

# Calculate and display the percentage of each hotel type
print("\nPercentage of Hotel Types:")
display(df['hotel'].value_counts(normalize=True) * 100)
```

```python
# Create a histogram for the 'lead_time' variable
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='lead_time', bins=50, kde=True) # Using 50 bins to see the distribution better
plt.title('Distribution of Lead Time')
plt.xlabel('Lead Time (Days)')
plt.ylabel('Number of Bookings')
plt.show()
```

```python
# #summary analysis for lead_time
# Use info() for the lead_time variable
print("Info of lead_time:")
df['lead_time'].info()

# Use describe() for the lead_time variable
print("\nDescription of lead_time:")
display(df['lead_time'].describe())
```

## 5. Bivariate Analysis

Business framing:  

Stakeholders often ask: “What drives cancellations?” or “Do longer stays mean higher revenue?” Bivariate analysis helps you uncover those kinds of relationships.

- Choose 2 relevant variable pairs (e.g., `lead_time` vs. `is_canceled`, or `adr` vs. `customer_type`)
- Use scatterplots, grouped bar plots, or boxplots to explore the relationships
- Interpret what these relationships could mean for the hotel business

```python
# Your code to analyze variable relationships (e.g., scatterplots, grouped bars) 🔧
```

```python
# Create a grouped bar plot comparing arrival_date_month and hotel
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='arrival_date_month', hue='hotel', order=['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']) # Order months chronologically
plt.title('Number of Bookings by Month and Hotel Type')
plt.xlabel('Arrival Month')
plt.ylabel('Number of Bookings')
plt.xticks(rotation=45, ha='right') # Rotate x-axis labels for better readability
plt.legend(title='Hotel Type')
plt.tight_layout() # Adjust layout to prevent labels overlapping
plt.show()
```

```python
# Create a box plot comparing is_canceled and lead_time
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='is_canceled', y='lead_time')
plt.title('Lead Time vs. Cancellation Status')
plt.xlabel('Is Canceled (0: Not Canceled, 1: Canceled)')
plt.ylabel('Lead Time (Days)')
plt.xticks([0, 1], ['Not Canceled', 'Canceled']) # Label x-axis ticks
plt.show()
```
